```python
Task 1_DS:Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable,
           such as the distribution of ages or genders in a population.
```

```python
 import matplotlib.pyplot as plt
genders = ['Male', 'Female', 'Other']
counts = [45, 55, 5] 


plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['blue', 'pink', 'purple'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender in Sample Population')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
ages = np.random.randint(18, 65, size=100) 


plt.figure(figsize=(8, 6))
plt.hist(ages, bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in Sample Population')
plt.grid(True)
plt.show()
```
